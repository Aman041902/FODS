```python

Q1 = dataset['chol'].quantile(0.25)
Q3 = dataset['chol'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define the bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
dataset_cleaned = dataset[(dataset['chol'] >= lower_bound) & (dataset['chol'] <= upper_bound)]

# dataset_cleaned now contains the dataset without outliers in the 'chol' column
```

```python

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(dataset_cleaned['chol'], bins=10, kde=True, color='lightgreen')
plt.title('Histogram of Cholesterol Levels')
plt.xlabel('Cholesterol (chol)')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.boxplot(y=dataset_cleaned['chol'], color='lightcoral')
plt.title('Box Plot of Cholesterol Levels')
plt.ylabel('Cholesterol (chol)')

plt.tight_layout()
plt.show()
```

```python

Q1 = dataset['trestbps'].quantile(0.25)
Q3 = dataset['trestbps'].quantile(0.75)


IQR = Q3 - Q1

# Define the bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


dataset_cleaned = dataset[(dataset['trestbps'] >= lower_bound) & (dataset['trestbps'] <= upper_bound)]
```

```python

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(dataset_cleaned['trestbps'], bins=10, kde=True, color='skyblue')
plt.title('Histogram of Resting Blood Pressure')
plt.xlabel('Resting Blood Pressure (trestbps)')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.boxplot(y=dataset_cleaned['trestbps'], color='salmon')
plt.title('Box Plot of Resting Blood Pressure')
plt.ylabel('Resting Blood Pressure (trestbps)')

plt.tight_layout()
plt.show()
```
